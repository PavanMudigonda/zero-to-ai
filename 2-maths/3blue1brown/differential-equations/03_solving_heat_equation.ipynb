# Chapter 3: Solving the Heat Equation## Separation of VariablesA powerful technique for solving PDEs!### The IdeaAssume the solution factors:$$T(x,t) = X(x) \cdot \Theta(t)$$Spatial part × Temporal part

```python
import numpy as npimport matplotlib.pyplot as pltimport seaborn as snsfrom matplotlib.animation import FuncAnimationfrom IPython.display import HTMLsns.set_style('whitegrid')plt.rcParams['figure.figsize'] = (14, 10)np.set_printoptions(precision=3, suppress=True)
```

## The MethodSubstitute T(x,t) = X(x)Θ(t) into heat equation:$$X(x)\Theta'(t) = \alpha X''(x)\Theta(t)$$Divide both sides by XΘ:$$\frac{\Theta'(t)}{\alpha\Theta(t)} = \frac{X''(x)}{X(x)} = -\lambda$$Both sides must equal the same constant!This gives us **two ODEs**:1. Θ'(t) = -αλΘ(t)2. X''(x) = -λX(x)

```python
def show_separation():    """Demonstrate separation of variables."""    x = np.linspace(0, np.pi, 100)    t_vals = [0, 0.5, 1.0, 2.0]        fig, axes = plt.subplots(2, 2, figsize=(14, 10))        for idx, t in enumerate(t_vals):        ax = axes[idx//2, idx%2]                # Solution: sum of modes        T = np.zeros_like(x)        for n in range(1, 6):            lam = n**2            X = np.sin(n * x)            Theta = np.exp(-lam * t)            T += X * Theta / n**2                ax.plot(x, T, 'b-', linewidth=2)        ax.set_title(f't = {t:.1f}', fontweight='bold')        ax.set_xlabel('x')        ax.set_ylabel('T(x,t)')        ax.grid(True, alpha=0.3)        ax.set_ylim(-0.5, 3)        plt.tight_layout()    plt.show()        print("Each mode decays exponentially in time!")    print("Higher frequencies decay faster.")show_separation()
```

## Boundary ConditionsFor a rod of length L with fixed temperatures at ends:- T(0, t) = 0- T(L, t) = 0This gives us: X(x) = sin(nπx/L)**Eigenvalues**: λₙ = (nπ/L)²### General Solution$$T(x,t) = \sum_{n=1}^{\infty} A_n \sin\left(\frac{n\pi x}{L}\right) e^{-\alpha(n\pi/L)^2 t}$$Each term is a **mode** that decays exponentially!

## Summary**Separation of variables** converts PDE → ODEs**Key steps**:1. Assume T(x,t) = X(x)Θ(t)2. Separate into two ODEs3. Apply boundary conditions4. Sum up all modes**Next**: Fourier series and complex exponentials!
