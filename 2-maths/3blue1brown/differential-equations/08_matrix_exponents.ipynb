## The Solution$$\vec{x}(t) = e^{At}\vec{x}(0)$$But what does e^(At) mean?### Matrix Exponential$$e^{At} = I + At + \frac{(At)^2}{2!} + \frac{(At)^3}{3!} + \cdots$$Like Taylor series, but with matrices!

```python
def matrix_exponential_demo():    """Demonstrate matrix exponential."""    from scipy.linalg import expm        # Simple rotation matrix    theta = np.pi / 4    A = np.array([[0, -1], [1, 0]]) * theta        print("Matrix A (rotation generator):")    print(A)    print("\nMatrix exponential e^A:")    print(expm(A))    print("\nThis is a rotation matrix!")        # Verify    c, s = np.cos(theta), np.sin(theta)    R = np.array([[c, -s], [s, c]])    print("\nExpected rotation:")    print(R)matrix_exponential_demo()
```
