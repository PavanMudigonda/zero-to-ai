### The MNIST Challenge

**Goal**: Recognize handwritten digits (0-9)

- 28×28 pixel images = 784 inputs
- Output: Which digit?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (16, 10)
np.random.seed(42)
```

## Network Structure

**Layers**:
1. Input Layer: 784 neurons
2. Hidden Layers: 16 neurons each
3. Output Layer: 10 neurons

Each neuron has an **activation** (0 to 1).

```python
# Visualize network
print("Network: 784 → 16 → 16 → 10")
print("Total parameters: ~13,000")
```

## Activation Function

Sigmoid: σ(x) = 1/(1 + e^(-x))

Squashes values to [0, 1]
