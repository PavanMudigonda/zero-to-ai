## Gradient Descent

Imagine finding the lowest point in a foggy valley:
1. Check which direction is steepest downhill
2. Take a small step
3. Repeat!

Update rule: w_new = w_old - η∇C

```python
# Simple gradient descent demo
import numpy as np
w = 4.0
for i in range(10):
    grad = 2*(w-2)  # derivative
    w = w - 0.3*grad
    print(f"Step {i}: w={w:.4f}")
print(f"Final: {w:.4f}")
```
