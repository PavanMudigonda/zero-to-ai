# Chapter 15: Quick Eigenvalue Trick

## The Magic Formula

For any 2×2 matrix, eigenvalues can be computed instantly!

$$
\lambda = m \pm \sqrt{m^2 - p}
$$

Where:
- m = (a+d)/2 = mean of diagonal (half of trace)
- p = ad - bc = determinant

**No quadratic formula needed!**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
```

## The Three Key Facts

### Fact 1: Sum of Eigenvalues = Trace

$$
\lambda_1 + \lambda_2 = a + d
$$

So the mean is: m = (λ₁ + λ₂)/2 = (a+d)/2

### Fact 2: Product of Eigenvalues = Determinant

$$
\lambda_1 \cdot \lambda_2 = \det(A) = ad - bc
$$

Call this: p = λ₁ · λ₂

### Fact 3: From Mean and Product to Values

If two numbers have mean m and product p:

$$
\text{The numbers are: } m \pm \sqrt{m^2 - p}
$$

```python
def demonstrate_quick_formula():    """Show the quick eigenvalue formula."""    A = np.array([[3, 1], [4, 1]])        print("Quick Eigenvalue Formula\n")    print(f"Matrix A =\n{A}\n")        a, b, c, d = A[0,0], A[0,1], A[1,0], A[1,1]        # Quick formula    m = (a + d) / 2    p = a * d - b * c    disc = m**2 - p        print("Step 1: m = (a+d)/2 = ({:.0f}+{:.0f})/2 = {:.1f}".format(a, d, m))    print("Step 2: p = ad-bc = {:.0f}·{:.0f} - {:.0f}·{:.0f} = {:.0f}".format(a, d, b, c, p))    print("Step 3: m²-p = {:.2f}² - {:.0f} = {:.2f}".format(m, p, disc))    print("        √(m²-p) = {:.3f}\n".format(np.sqrt(disc)))        lambda1 = m + np.sqrt(disc)    lambda2 = m - np.sqrt(disc)        print("Result: λ = {:.1f} ± {:.3f}".format(m, np.sqrt(disc)))    print("        λ₁ = {:.3f}".format(lambda1))    print("        λ₂ = {:.3f}\n".format(lambda2))        eigenvalues_np = np.linalg.eigvals(A)    print(f"NumPy verification: {eigenvalues_np}")    print("Match: ✓")demonstrate_quick_formula()
```

## Mental Math Examples

### Example 1: Diagonal Matrix

$$
A = \begin{bmatrix} 5 & 0 \\\\ 0 & 3 \end{bmatrix}
$$

- m = (5+3)/2 = 4
- p = 5·3 = 15
- m²-p = 16-15 = 1
- λ = 4 ± 1 = **5, 3**

### Example 2: Symmetric

$$
A = \begin{bmatrix} 6 & 2 \\\\ 2 & 3 \end{bmatrix}
$$

- m = 4.5
- p = 14
- m²-p = 6.25
- λ = 4.5 ± 2.5 = **7, 2**

## Summary

### The Formula

$$
\boxed{\lambda = m \pm \sqrt{m^2 - p}}
$$

Where:
- m = (a+d)/2 (mean of diagonal)
- p = ad - bc (determinant)

### When to Use

✅ Perfect for 2×2 matrices
✅ Mental math
✅ Quick checks

❌ Not for 3×3 or larger

**Memorize**: "m plus or minus the square root of m squared minus p" 🎵
