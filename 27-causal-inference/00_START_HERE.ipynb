## Correlation vs. Causation

Standard ML finds correlations: *"users who see this ad also buy product X"*.

Causal inference asks: *"Does showing this ad **cause** users to buy product X?"*

This distinction matters enormously for decisions, policy, and AI alignment.

## Key Concepts

| Concept | Description |
|---------|-------------|
| **Potential outcomes** | Y(1), Y(0) — what would have happened |
| **ATE** | Average Treatment Effect |
| **DAG** | Directed Acyclic Graph — causal structure |
| **Confounders** | Variables affecting both treatment and outcome |
| **Propensity score** | P(treatment | covariates) — for matching |
| **IV** | Instrumental Variables — for unmeasured confounders |
| **Diff-in-diff** | Before/after comparison with control group |
