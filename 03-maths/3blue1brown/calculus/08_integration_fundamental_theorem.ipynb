## Integrals: The Inverse of Derivatives

Integration accumulates infinitesimal pieces.

### The Fundamental Theorem of Calculus

$$
\int_a^b f'(x)\,dx = f(b) - f(a)
$$

**Derivatives and integrals are inverses!**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 10)
```

## Area Under a Curve

The integral computes area:

$$
A = \int_a^b f(x)\,dx
$$

We approximate with thin rectangles and take the limit.

```python
def riemann_sum():
    def f(x):
        return x**2
    
    a, b = 0, 3
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for idx, n in enumerate([5, 20, 100]):
        ax = axes[idx]
        x = np.linspace(a, b, 1000)
        ax.plot(x, f(x), 'b-', linewidth=2)
        
        # Rectangles
        x_rect = np.linspace(a, b, n)
        dx = (b - a) / n
        for xi in x_rect[:-1]:
            height = f(xi)
            rect = plt.Rectangle((xi, 0), dx, height, 
                                facecolor='lightblue', edgecolor='blue', alpha=0.5)
            ax.add_patch(rect)
        
        area = sum(f(xi) * dx for xi in x_rect[:-1])
        ax.set_title(f'n={n}, Area≈{area:.2f}', fontweight='bold')
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    print(f"Exact: ∫₀³ x² dx = [x³/3]₀³ = 9")

riemann_sum()
```
