## Attention Mechanism

Words pay different amounts of attention to each other!

Example: 'The cat sat on the mat'
- 'sat' attends to 'cat' (subject)
- 'sat' attends to 'mat' (object)

## Query, Key, Value

Each word produces 3 vectors:
- **Query**: What am I looking for?
- **Key**: What do I contain?
- **Value**: What do I communicate?

Attention score = Query · Key
