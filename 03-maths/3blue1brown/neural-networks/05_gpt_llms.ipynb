## Tokenization

Text → tokens (words/pieces)

Each token → vector embedding

Similar words have similar embeddings!

```python
# Simple tokenization
text = "The quick brown fox"
tokens = text.split()
print("Tokens:", tokens)
print(f"Count: {len(tokens)}")
```
