## How Diffusion Works

### Training
1. Take real image
2. Add noise gradually  
3. Train network to predict noise

### Generation
1. Start with pure noise
2. Denoise step by step
3. Get realistic image!

**Applications**: DALL-E, Stable Diffusion, Midjourney
